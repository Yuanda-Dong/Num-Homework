```python
import numpy as np
import matplotlib.pyplot as pt

k = np.random.random() * 4 + 3
abs_err_values = []
truncation_err_values = []
rounding_err_values = []
n_values = [2**n_exp for n_exp in range(5, 20)]
def f(x):
    return np.sin(k * x)

def f_prime(x):
    return -k**2*np.sin(k*x)

def f_prime_app(x,h):
    return (f(x+h)-2*f(x)+f(x-h))/h**2

for n in n_values:
    h = 1/(n-1)
    x = np.linspace(0, 1, n)
    eps = np.finfo(np.float64).eps
    diffy = f_prime(x[1:-1]) 
    for i in range(diffy.shape[0]):
        diffy[i] -= f_prime_app(x[i+1],h)
    abs_err_values.append(np.max(np.abs(diffy))) 
    truncation_err_values.append(k**4*h**2/12)
    rounding_err_values.append(4*k*eps/h**2)
abs_err_values = np.array(abs_err_values)
truncation_err_values = np.array(truncation_err_values)
rounding_err_values = np.array(rounding_err_values)
# ------------ plotting code below, no need to change
h_values = []
for n in n_values:
    h_values.append(1 / (n - 1))
h_values = np.array(h_values)

pt.xlabel(r"$h$")
pt.loglog(h_values, truncation_err_values + rounding_err_values, label=r"Predicted Error Bound")
pt.loglog(h_values, abs_err_values, label="Computed Absolute Error")
pt.legend(loc='lower right')
pt.show()
```
